```python
# import libraries 
from config import username, password
import pandas as pd
import psycopg2

# set # of columns and rows to display
pd.set_option('display.max_columns', None)
pd.options.display.max_colwidth = None

# set width and height of ipython cell
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
display(HTML("<style>div.output_scroll { height: auto !important; max-height:500px; }</style>"))

# build a connection to ctti-clinicaltrials.org database
conn = psycopg2.connect(
    user = username,
    password = password,
    host = "aact-db.ctti-clinicaltrials.org",
    port = "5432",
    database = "aact")
```

```python
# query columns from information_schema
q = """
    select * from information_schema.columns limit 10;
"""

cols = pd.read_sql(q, conn)
cols
```

```python
# query tables from information_schema
q = """
    select * from information_schema.tables limit 10;
"""

tables = pd.read_sql(q, conn)
tables
```

```python
# query views from information_schema
q = """
    select * from information_schema.views limit 10;
"""

views = pd.read_sql(q, conn)
views
```
