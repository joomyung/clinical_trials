```python
from config import username, password
import pandas as pd
import psycopg2
```

```python
def execute(query):

    conn = psycopg2.connect(
    user=username,
    password=password,
    host="aact-db.ctti-clinicaltrials.org",
    port='5432',
    database="aact"
    )

    df=pd.read_sql(query, conn)

    return df
```

```python
q = "select * from information_schema.tables"
tables = execute(q)
```
